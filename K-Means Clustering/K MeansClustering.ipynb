```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("C:\\Users\\Rohit\\PycharmProjects\\Pycharm Projects\\K MeansClustering\\income.csv")
df
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1.Age,df1['Income($)'],color="green",marker = "+")
plt.scatter(df2.Age,df2['Income($)'],color="blue",marker = ".")
plt.scatter(df3.Age,df3['Income($)'],color="red",marker = "*")
```
